```python
with open("day_14.txt", "r") as f:
    inp = f.readlines()
```

```python
import re
import numpy as np

class Memory():
    def __init__(self):
        self.memory = {}
        
    def register(self, address, value):
        self.memory[address] = value

    def get_sum(self):
        return sum([s.value for s in self.memory.values()])
    
class Binary():
    int_size = 36
    
    def __init__(self, line, memory):
        res = re.match('(?:(mask)|(mem)\[(\d+)\]) = (.*)\n?', line)
        mask, mem, address, value = res.group(1, 2, 3, 4)
        
        if not mask is None:
            self.set_mask(value)
        else:
            self.address = address
            self.set_value(int(value))
            memory.register(self.address, self)
            
    def set_value(self, value):
        self.value = np.empty((self.int_size), dtype=str)
        bin_value = bin(value)[2:]
        bin_value = ''.join(['0']*(self.int_size - len(bin_value))) + bin_value
        for i in range(self.int_size):
            if self.mask[i] == 'X':
                self.value[i] = str(bin_value[i])
            else:
                self.value[i] = str(self.mask[i])
        self.value = int(''.join(self.value.tolist()), 2)
        
    def __str__(self):
        try:
            return str(self.value)
        except AttributeError:
            return str(self.mask)
        
    def is_mask(self):
        try:
            return not self.value
        except AttributeError:
            return True
    
    @classmethod
    def set_mask(cls, mask):
        cls.mask = mask
```

```python
memory = Memory()
mask = Binary(inp[0], memory)
for line in inp:
    a = Binary(line, memory)
    if not a.is_mask():
        print(a.value)
print(memory.get_sum())
```
